# Unit 1 — Your first (real) agent

Two things to feel:
1. `instructions` shape the agent's personality for every call.
2. `run` gives you the whole response; `run(stream=True)` yields chunks as they arrive.

```python
from dotenv import load_dotenv
load_dotenv(dotenv_path="../.env")

from agent_framework.openai import OpenAIChatCompletionClient

agent = OpenAIChatCompletionClient().as_agent(
    name="PirateAgent",
    instructions="You are a terse pirate. Reply in two short sentences, using pirate vocabulary.",
)
```

## Non-streaming

`await agent.run(...)` returns an `AgentResponse`. `.text` is the full reply.

```python
result = await agent.run("What is the best way to learn programming?")
print(result.text)
```

## Streaming

`agent.run(..., stream=True)` is an **async iterator**. Each `chunk` carries a small piece of `.text`. This is what ChatGPT does in its web UI.

In a notebook the effect is subtle — Jupyter batches output. Run the `.py` file in a terminal to see each token appear live.

```python
import sys

async for chunk in agent.run("What is the best way to learn programming?", stream=True):
    if chunk.text:
        print(chunk.text, end="", flush=True)
        sys.stdout.flush()
print()
```

## Tinker

Change the `instructions` to something totally different. The agent's identity changes instantly — no code changes, just text.

```python
doctor = OpenAIChatCompletionClient().as_agent(
    name="Doctor",
    instructions="You are a cautious medical doctor. Always ask a clarifying question before answering.",
)
result = await doctor.run("I have a headache.")
print(result.text)
```
