# Unit 0 — Sanity check

One goal: prove your env works.

If every cell in this notebook runs and the last one prints a greeting, you're ready for Unit 1.

## Step 1: load environment variables

`OpenAIChatCompletionClient()` reads `OPENAI_API_KEY`, `OPENAI_BASE_URL`, and `OPENAI_CHAT_COMPLETION_MODEL` from the environment. `load_dotenv()` pulls them from `../.env`.

```python
from dotenv import load_dotenv
load_dotenv(dotenv_path="../.env")

import os
assert os.environ.get("OPENAI_API_KEY"), "OPENAI_API_KEY missing — check ../.env"
assert os.environ.get("OPENAI_CHAT_COMPLETION_MODEL"), "OPENAI_CHAT_COMPLETION_MODEL missing — check ../.env"
print("env loaded, model:", os.environ.get("OPENAI_CHAT_COMPLETION_MODEL"))
```

## Step 2: create the agent

`as_agent()` is a helper on the client. It returns an `Agent` ready to run. `instructions` is the 'system prompt' baked into this agent for every call.

```python
from agent_framework.openai import OpenAIChatCompletionClient

agent = OpenAIChatCompletionClient().as_agent(
    name="HelloAgent",
    instructions="You are a friendly assistant. Keep answers to one sentence.",
)
agent
```

## Step 3: run it

In a notebook, Jupyter already has an event loop, so we `await` directly instead of calling `asyncio.run(...)`.

```python
result = await agent.run("Say hello in one short sentence.")
print(result.text)
```

## Tinker

Change the instructions and re-run. Try: `"You are Shakespeare. Reply in iambic pentameter."`

```python
shakespeare = OpenAIChatCompletionClient().as_agent(
    name="Bard",
    instructions="You are Shakespeare. Reply in iambic pentameter, one short line only.",
)
result = await shakespeare.run("Greet me.")
print(result.text)
```
